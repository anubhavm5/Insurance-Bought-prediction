```python
import pandas as pd
```

```python
df=pd.read_csv("Insurance_data.csv")
```

```python
df.head()
```

```python
x=df.drop(["bought_insurance"],axis=1)
```

```python
x
```

```python
y=df.bought_insurance
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Insurance Policy Analyzation")
plt.xlabel("Age Of Customer",color="black")
plt.ylabel("Bought Insurance",color="black")
plt.scatter(x,y,color="Red")
```

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[60]])
```

```python
age = int(input("Enter The Age: "))
pred = model.predict([[age]])
if pred[0]==0:
    print("Not Bought Insurance")
else:
    print("Bought Insurance")
    
```
